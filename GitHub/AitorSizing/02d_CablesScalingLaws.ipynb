### $r=I^{(1/1.5)}$

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

path='./Data/'
df = pd.read_csv(path+'ScalingLawsCables.csv',sep=';',encoding = 'unicode_escape')


f, (ax1) = plt.subplots(1, 1, figsize=(12,6))


#I=np.linspace(df['Imax[A]'].min(),df['Imax[A]'].max())

ax1.plot(df['Imax[A]'], df['diameter[mm]'],'o')
ax1.plot(I_ref, r_ref,'^g')


r_ref=0.51#[mm]
I_ref=3.5#[A]
M_ref=0.001830367#[kg]

I=df['Imax[A]']

r=r_ref*(I/I_ref)**(1/1.5)

ax1.plot(I,r)

#Here we plot
#Labelling
#ax1.loglog()
ax1.grid(True,which="both",ls="-")
ax1.legend()
ax1.set_title('r=I^(1/1.5) ')
ax1.set_xlabel('I [A]')
ax1.set_ylabel('r [mm]')
```
