## Momentum theory

The design of a drone's rotor constitutes the cornerstone for the aerodynamics of the vehicle. For this purpose, a matter of special importance is to know the three basic functions that a rotor must provide:
- Generation of lift
- Generation of propulsive forces for the forward flight
- Generation of forces to control the flight position and attitude.

These forces have to be applied to control the drone in the different flight scenarios like:
- Hover
- Climb / Descend
- Forward / Backward flight

and any combination of the scenarios above.

In 1865, W.J.M. Rankine *et al* developed the momentum theory for marine propellers. This theory applies the basic conservation laws of fluid mechanics ( conservation of mass, momentum and energy) to the rotor to estimate the rotor performance.

### Assumptions

Hover:
 - no forward speed
 - no vertical speed
 - axisymmetrical flow field
 - wake boundary with the flow outside this boundary.
 
Rotor:
 - rotor modeled as an actuator disk which adds momentum and energy to the flow.
 - incompressible, steady and irrotational flow.

## Equations

### Conservation of Mass

The mass flux through the disk is 

$\displaystyle \dot m = \rho A v$ 

and by conservation of mass, the mass flux is constant all along the wake.

### Momentum and energy equations

Momentum conservation equates the rotor force $T$ to the rate of change of momentum flowing out at station 3: 

$ \displaystyle T=\dot m w$

According to the energy conservation, the work done per unit time (power) by the rotor is equal to the energy rate of change in the fluid:

$ \displaystyle T v_i=\frac{1}{2} \dot m w^2$

Eliminating $T/\dot m$ from the momentum and energy conservation equation gives that the far wake induce velocity is twice the velocity at the disk:

$\displaystyle w=2v_h$

### Ideal power and disk loading

Momentum theory relates the rotor thrust and the induced velocity at the rotor disk by :

$\displaystyle T= \dot m w= 2 \rho A v^2$ 

The induced velocity is: 

$\displaystyle v=\sqrt{\frac{T}{2 \rho A}}$

The induced power loss for hover is (Power consumed=Energy rate flow out-Energy rate flow in
):

$\displaystyle P=T v= T \sqrt{\frac{T}{2 \rho A}}$

Disk loading is defined as the ration of the thrust by the disk area:
 $\displaystyle DL= \frac{T}{A}$
 
Another form to express the induced velocity can be by the disk loading:
$\displaystyle v= \sqrt{\frac{DL}{2 \rho}}$
